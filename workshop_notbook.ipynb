```python
import pandas as pd

modalites = ['TECO', 'Box', 'CRTD', 'REL', 'Notif', 'box', 'CNF', 'PCNF', 'tECO']
df_cleaned['Completion'] = df_cleaned['Completion'].fillna(pd.Series(modalites).sample(n=1, replace=True).iloc[0])
for col in columns_to_check:
    print(f"{col} : {df_cleaned[col].nunique()} valeurs uniques")
    print(df_cleaned[col].unique())
    print("-" * 50)
```

```python

import re

def clean_equipment(value):
    if isinstance(value, str):
        value = value.lower().strip().replace(" / ", ", ").replace("  ", " ")

        if re.match(r"^tr-\d{2}$", value):
            return "tr-series"
        value = value.replace("sts compressor's box", "sts compressor box")
        value = value.replace("all equipments", "all equipment")

        return value
    return value
df_cleaned['ID: Activity/ Equipment '] = df_cleaned['ID: Activity/ Equipment '].apply(clean_equipment)


df_cleaned['ID: Activity/ Equipment '] = df_cleaned['ID: Activity/ Equipment '].fillna("Inconnu")
print(df_cleaned['ID: Activity/ Equipment '].unique())

```

```python
def clean_status(value):
    if isinstance(value, str):
        value = value.lower().strip()

        if value == 'closed':
            return 'closed'
        elif value == 'active':
            return 'active'

        return value
    return value
df_cleaned['Status'] = df_cleaned['Status'].apply(clean_status)


print("Valeurs uniques aprÃ¨s rÃ©duction dans 'Status' :", df_cleaned["Status"].nunique())
print(df_cleaned["Status"].unique())
```
