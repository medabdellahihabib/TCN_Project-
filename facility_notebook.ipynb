#2.Data processing

```python
df.drop(columns=['WCR', 'Completion'], inplace=True, errors='ignore')


nan_percentage = (df_cleaned.isna().sum() / len(df_cleaned)) * 100
print(nan_percentage)

```

```python
colonnes_a_traiter = [
    "Description", "Work center / Performed By", 'Name of Persons Allocated for the Activity ',
    "No. of Manpower", "Status", "Supervisor", "Shift", "W",
    "Date", "Month", "Year", "Duration time Format hh:mm:ss"
]

df_cleaned = df_cleaned.dropna(subset=colonnes_a_traiter)
df_cleaned = df_cleaned.dropna(axis=1, how='all')

nan_percentage = (df_cleaned.isna().sum() / len(df_cleaned)) * 100
print(nan_percentage)
```

```python

print(" Nombre de valeurs uniques par colonne :\n")
for col in df_cleaned.columns:
    valeurs_uniques = df_cleaned[col].nunique(dropna=True)
    print(f"✅ {col}: {valeurs_uniques} valeurs uniques")
```

```python
df_cleaned = df_cleaned.drop(columns=['Operation '])
```
